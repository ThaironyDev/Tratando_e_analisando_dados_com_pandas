## Exercício 2

```python
import pandas as pd
```

>Carregando dados

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url, sep=',')
```

>Verificando se o dataframe possui dados nulos.

```python
dados.isnull().sum()
```

>Substituindo os valores nulos por zero.


```python
dados.fillna(0, inplace=True)
```

> Removendo os alunos 'Alice' e 'Carlos' que não fazem mais parte da turma.

```python
selecao = dados.query('Nome == "Alice" | Nome == "Carlos"').index
dados.drop(selecao, axis=0, inplace=True)
```

>Selecionando apenas os alunos aprovados.

```python
aprovados = dados.Aprovado == True
alunos_aprovados = dados[aprovados]
```

> Salvando um arquivo csv com os alunos aprovados.

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False, sep=';')
```

> Corrigindo as notas do alunos aprovados substitiuindo as notas 7.0 por 8.0 .

```python
alunos_aprovados = pd.read_csv('alunos_aprovados.csv', sep=';')
alunos_aprovados.head()
```

```python
#alunos_aprovados.loc[alunos_aprovados['Notas'] == 7.0] = 8.0
alunos_aprovados.Notas.replace(7, 8, inplace=True)
```

```python
alunos_aprovados.head()
```

>Salvando novamente o arquivo csv com os alunos_aprovados e notas corrigidas.

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False, sep=';')
```

##Exercício 3

>Carregando arquivo

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url, sep=',')
```

>Criando uma nova coluna chamada Pontos_extras que corresponde a 40% da nota atual de cada aluno.

```python
dados.fillna(0, inplace=True)
selecao = dados.query('["Alice", "Carlos"] in Nome').index
dados.drop(selecao, axis=0, inplace=True)
dados['Pontos_extras'] = dados['Notas'].apply(lambda x: x * .4)
dados.head()
```

>Criando um coluna chamada Notas_finais que é a somatória das notas de cada aluna e seus pontos extras.

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados.head()
```

>Criando uma coluna chamada Aprovado_final que corresponde a situação atual do aluno que não conseguiu pontução suficiente para ser aprovado e agora devido aos pontos extras pode ser aprovado.

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x: True if x >= 6 else False)
dados.head()
```

>Selecionando os alunos que não tinham sido aprovados anteriormente mas agora devido aos pontos extras conseguiram ser aprovados.

```python
selecao = (dados['Aprovado'] == False) & (dados['Aprovado_final'] == True)
dados[selecao]
```
