> Usando o método groupby() para agrupar os imóvies por tipo.

```python
pd.__version__ # Verificando a versão do Pandas.
```

>Para esse objetivo estamos usando o mean() como um método de agregação, ou seja, após agruparmos os dados
de acordo com a coluna "Tipo" calcularemos a média da coluna "Valor" para cada tipo de imóvel.
OBS: caso estejamos usando uma versão do pandas inferior a 2 notaremos que a média será calculada exclusivamente para
as colunas de valor númerico, caso cotrário a partir da versão 2 o pandas irá tentar calcular a média para todas as colunas gerando um erro,
para evitar esse erro podemos passar um parâmetro para o método mean(numeric_only = True) que diz ao panda para calcular a média
apenas para as colunas númericas.

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

>Note que na célula anterior apenas fizemos o agrupamento mas não selecionamos uma coluna especifica para calcular a
média, agora selecionamos a coluna "Valor" que nos interessa para este fim. Diante dessa indicação da coluna de interesse
não temos a necessidade de utilizar o parâmetro numeric_only=True pois conhecemos de antemão o tipo de dado da coluna de interesse.
O retorno é uma Series que tem como indices o tipo de cada imovel e como valor a média do aluguel.

```python
dados.groupby('Tipo')['Valor'].mean()
```

> Agora vamos mudar o formato de retorno da nossa pesquisa modificando nosso código para retornar um dataframe. Note que ao
especificar a coluna de interesse usamos um par de colchetes aninhados isso corresponde a criaçãode um dataframe que nada mais é que
um grupo de Series(colunas), depois usamos o método sort_values() para ordena os valores do menor para o maior.

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

> Agora usando o método plot do pandas vamos plotar um gráfico usando nosso dataframe da média dos valores por tipo. Os parâmetros usados para este fim são:

kind = 'barh' -> Especificando o tipo de gráfico como barras horizontais, caso o h não seja informado as barras serão verticais.

figsize = (10, 6) -> Especifica o tamanho deste gráfico

color = 'purple' -> Especifica a cor das barras

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(10, 6), color='purple')
```

## Removendo os imóvies comerciais

>Verificando os tipos únicos de imóveis na base de dados usando o método unique()

```python
dados.Tipo.unique()
```

> Aqui estamos criando dois datasets um para imoveis comerciais e outro para residênciais. Para isso usamos o dataframe original dados que contém todos os dados que precisamos e selecionamos uma coluna "Tipo" como pivo da nossa lógica de seleção, ou seja, estamos instanciando um novo dataframe desde que obedeça a logica de seleção, a coluna "Tipo" que neste caso inicia-se com uma negação ~ do resultado da seleção da coluna em questão onde os valores da lista categoria_residencial estejam arrolados.

```python
categoria_residencial = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila', 'Flat', 'Loft', 'Studio']
imoveis_comerciais = dados[~dados['Tipo'].isin(categoria_residencial)]
imoveis_residenciais = dados[~dados['Tipo'].isin(imoveis_comerciais['Tipo'])]
imoveis_residenciais.Tipo.unique()
```

>Podemos usar o método query(), para este mesmo fim. O método query aceita uma condição para retornar um resultado o @imoveis_residenciais é usando para referenciar a variável de interesse, então para referenciar variais no método query precisamos usar o @.

```python
imoveis_comerciais = dados.query('@categoria_residencial not in Tipo')
imoveis_residenciais = dados.query('@categoria_residencial in Tipo')
```

```python
df = imoveis_residenciais
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo
```

```python
df_preco_tipo.plot(kind='barh', figsize=(10, 6), color='green')
```

## Qual o percentual de cada tipo de imóvel na nossa base de dados?

>Contando quantas vezes um valor aparece em uma coluna usando o método value_counts().
O retorno desse método é uma Series com indices nomeados cujo valor é a soma de todas as ocorrências registradas, para que a apresentação seja um valor percentual usamos o parâmetro normalize=True, teremos os retorno de um percentual para todas as ocorrencias registradas nessa coluna sobre cada valor que está na mesma.

```python
df.Tipo.value_counts(normalize=True)
```

>Convertendo um Series em um Dataframe usano o metodo to_frame().
Renomeamos a coluna Tipos para Percentuais usando o método rename(), passando como parâmetro para columns um dicionario que tem como chave o nome antigo da coluna e valor o novo nome, o parâmetro inplace serve para indicar que essa mudança deve ser definitiva nesta variável.

```python
df_exemplo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_exemplo.rename(columns={'Tipo': 'Percentuais'}, inplace=True)
df_exemplo
```

## Criar colunas numéricas

##Criando colunas numéricas

>Importanto novamente os dados pois vamos considerar todas as informações para esta atividade.

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
```

>Criando uma coluna no dataframe e atribuindo a ela o resultado da somatória da coluna Valor e Condomínio.



```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

>Criando uma coluna uma coluna que vai guardar os gastos anuais de todos os imóvies, para isso devemos considerar os valores mensais já calculados na coluna Valor_por_mes multiplicados por 12 e somarmos esse resultado aos valores da coluna IPTU.

```python
dados['Valor_por_ano'] = (dados['Valor_por_mes'] * 12) + dados['IPTU']
dados.head()
```

##Criando colunas categóricas

>Concatenando as informações da coluna Descrição, além de alterar o tipo das colunas Quartos e Vagas para str, pois não é possível concatenar strings com numéricos.

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
 dados['Quartos'].astype(str) + ' quarto(s) e ' + \
 dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

> Criação da coluna Possui_Suite que deve conter um valor booleano. Para isso vamos avaliar se a coluna Suites possui um valor maior que zero. A utilização do metodo apply() em conjunto com uma função lambda garante o restorno esperado já que a função será executada para cada valor da coluna Suite. Veremos portanto se cada valor da conluna Suite repesentada por 'a' é > 0 para isso usamos a sintaxe do operador ternário que retorna um valor diferente caso o retorno da condição seja True ou False.

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda a: 'Sim' if a > 0 else 'Não')
dados.head()
```
